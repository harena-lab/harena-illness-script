```python
import pandas as pd
import json

# Read CSV file
csv_file_path = 'test_data_info_no_short.csv'
csv_data = pd.read_csv(csv_file_path)

# Read JSONL file
jsonl_file_path = '../teste-progresso/04-ml/annotations-medical_specialist-dpoc-full.jsonl'
jsonl_data = []
with open(jsonl_file_path, 'r',encoding='utf-8') as file:
    for line in file:
        jsonl_data.append(json.loads(line))

# Create a dictionary from jsonl_data with doc_id as key and text as value
jsonl_text_dict = {item['doc_id']: item['text'] for item in jsonl_data}

# Replace the text column in csv_data with the text from jsonl_data
csv_data['text'] = csv_data['doc_id'].map(jsonl_text_dict)

# Save the updated csv_data to a new CSV file
csv_data.to_csv(csv_file_path, index=False)
```
